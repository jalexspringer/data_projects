# Read the data 

```python
import pandas as pd
import numpy as np
import seaborn
import matplotlib.pyplot as plt
%matplotlib inline
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
```

# Cleaning. All the cleaning.

### Removing all rows with no respondent ID

```python
star_wars = star_wars[star_wars["RespondentID"].notnull()]
```

### Yes or No answers - change to Boolean

```python
yes_no = {
    "Yes" : True,
    "No" : False
}

star_wars["Have you seen any of the 6 films in the Star Wars franchise?"] = star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].map(yes_no)
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] = star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].map(yes_no)                                                      
```

### Rename columns to make the set easier to work with

```python
col_change_dict = {
    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
    "Unnamed: 4": "seen_2",
    "Unnamed: 5": "seen_3",
    "Unnamed: 6": "seen_4",
    "Unnamed: 7": "seen_5",
    "Unnamed: 8": "seen_6",
    "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "fav_1",
    "Unnamed: 10": "fav_2",
    "Unnamed: 11": "fav_3",
    "Unnamed: 12": "fav_4",
    "Unnamed: 13": "fav_5",
    "Unnamed: 14": "fav_6",
    "Please state whether you view the following characters favorably, unfavorably, or are unfamiliar with him/her.":"Han Solo"
}
char_list = ["Luke Skywalker","Princess Leia Organa","Anakin Skywalker","Obi Wan Kenobi","Emperor Palpatine","Darth Vader","Lando Calrissian","Boba Fett","C-3P0","R2 D2","Jar Jar Binks", "Padme Amidala", "Yoda"]
counter=16
for k in char_list:
    col_change_dict["Unnamed: {}".format(counter)] = k
    counter += 1
star_wars = star_wars.rename(columns=col_change_dict)
```

### Change "seen" values to Boolean

```python
remap_seen = {
    "Star Wars: Episode I  The Phantom Menace": True,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True,
    np.nan: False
}
star_wars.columns[3:9].fillna("None")
for i in star_wars.columns[3:9]:
    star_wars[i] = star_wars[i].map(remap_seen)

star_wars
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

## What is the people's favorite?
Lower is better - rankings 1 through 5 averaged.

```python
star_wars[star_wars.columns[9:15]].mean().plot(kind="bar", title="Star Wars Favorites Average", figsize=(12,8))
```

### Episode V is the winner
Unsurprisingly (and as it should be) the original three films have received the best rankings.

## Which of the films have been most widely seen?

```python
star_wars[star_wars.columns[3:9]].sum().plot(kind="bar", title="Star Wars Seen Films Average", figsize=(12,8))
```

### Episode V wins again
There is a pretty clear relationship here between the series. Episode IV - VI attracted increasing views (slight drop for Return of the Jedi), while Episode I disappointed fans and viewership dropped.

# Segmenting the data

## Do Trekkies have different opinions?
Exploring fans to see if Star Trek fans take a different view have different preferences in their sci-fi choices.

```python
trek_fans = star_wars[star_wars["Do you consider yourself to be a fan of the Star Trek franchise?"] == "Yes"]
trek_not_fans = star_wars[star_wars["Do you consider yourself to be a fan of the Star Trek franchise?"] == "No"]
trek_fans[trek_fans.columns[9:15]].mean().plot(kind="bar", title="Star Wars Favorites Average - Trekkies", figsize=(10,6))
```

```python
trek_not_fans[trek_fans.columns[3:9]].sum().plot(kind="bar", title="Star Wars Seen Films Average - Trekkies", figsize=(10,6))
```

```python
trek_not_fans[trek_not_fans.columns[9:15]].mean().plot(kind="bar", title="Star Wars Favorites Average - Non-Trekkies", figsize=(10,6))
```

```python
trek_not_fans[trek_not_fans.columns[3:9]].sum().plot(kind="bar", title="Star Wars Seen Average - Non-Trekkies", figsize=(10,6))
```

## Trekkies are great predictors of general SW belief trends
As if this surprised anyone! Trekkies clearly have great taste.

## Do men and women agree?


```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
males[males.columns[9:15]].mean().plot("bar", title="Star Wars Favorites Average - Men", figsize=(10,6))
```

```python
males[males.columns[3:9]].sum().plot("bar", title="Star Wars Seen Average - Men", figsize=(10,6))
```

```python
females[females.columns[9:15]].mean().plot("bar", title="Star Wars Favorites Average - Women", figsize=(10,6))
```

```python
females[females.columns[3:9]].sum().plot("bar", title="Star Wars Seen Average - Women", figsize=(10,6))
```

### Women less willing to see bad movies than men
While the rankings of each movie are in line with the general population, it appears that less women have see the entire series. This suggests a greater willingness to pass on movies if they are bad, even if they have the Star Wars name.

# Characters
## How are the characters ranked?

```python
star_wars[star_wars.columns[15:29]]
```

```python
char_sentiments = star_wars.dropna(axis=0, subset=[star_wars.columns[15:29]])
unique_sentiments = char_sentiments[char_sentiments.columns[15]].unique()
counter = 1
sentiment_map = {}
for i in unique_sentiments:
    sentiment_map[i] = counter
    counter += 1
sentiment_map['Unfamiliar (N/A)'] = np.nan
sentiment_map['Very unfavorably'] = 5
```

```python
for i in char_sentiments.columns[15:29]:
    char_sentiments[i] = char_sentiments[i].map(sentiment_map)
```

```python
char_sentiments[char_sentiments.columns[15:29]].mean().plot("bar", title="Star Wars Favorite Charactar Average - Lower is better", figsize=(10,6))
```

### Original heroes and comedic relief capture the hearts
Han Solo, Yoda, R2-D2 are among the favorites across the board. And now for the most important question of all...

# Who shot first?

```python
who_shot = star_wars[star_wars.columns[29]].dropna()
who_shot.value_counts().plot("bar")
```

### Clearly the survey takers are well informed
Episode V is considered the best and Han shot first. And women are less likely to blindly attend bad movies.
